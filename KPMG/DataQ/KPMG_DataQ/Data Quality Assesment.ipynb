# Downloading updated data

```python

import os
import urllib

DOWNLOAD_ROOT = 'https://s3-ap-southeast-2.amazonaws.com/insidesherpa-assets/vinternship_modules/kpmg_data_analytics/KPMG_VI_New_raw_data_update_final.xlsx'
DATA_PATH = os.path.join("datasets","customers")
FILE_NAME ='KPMG_VI_New_raw_data_update_final.xlsx'

def fetch_data(housing_url = DOWNLOAD_ROOT ,data_path = DATA_PATH,file_name = FILE_NAME): 
    #download data from a url into a local directory for later access
    os.makedirs(data_path, exist_ok = True)
    filepath= os.path.join(data_path,file_name)
    urllib.request.urlretrieve(housing_url, filepath)
    
    
```

```python
#function 
import pandas as pd

fetch_data()
sheets= ['Transactions','NewCustomerList','CustomerDemographic','CustomerAddress']

def load_data(data_path = DATA_PATH,file_name = FILE_NAME):
    #Return dict of dataframes for each sheet
    di = pd.read_excel (data_path+'/'+file_name,sheet_name=sheets)
    return di
    
```

```python
full_data=load_data()
```
